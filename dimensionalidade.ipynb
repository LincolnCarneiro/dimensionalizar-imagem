```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("/content/tangerina.webp")

plt.imshow(img)
plt.show()

print(img.shape)
```

```python
# gerando as primitivas da imagem

print(img[400][400][2]) # exemplo de um pxel da imagem

img_r = img.copy()
img_g = img.copy()
img_b = img.copy()
img_r[:, :, 1:2] = 0
img_g[:, :, 0:1] = 0
img_b[:, :, 0:-1] = 0

# transformando em escala de cinza

img_gray = img.copy()
gray_scale = img_gray[:, :, 0] * 0.30 + img_gray[:, :, 1] * 0.59 + img_gray[:, :, 2] * 0.11

img_gray[:,:,0] = gray_scale
img_gray[:,:,1] = gray_scale
img_gray[:,:,2] = gray_scale

plt.imshow(img_gray)
plt.show()
```

```python
# transformando em binarizado

treshold = 127

img_bin = img_gray.copy()

img_bin[img_bin < treshold] = 0
img_bin[img_bin >= treshold] = 255

plt.imshow(img_bin)
plt.show()
```
